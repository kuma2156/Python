```python
# 나이와 키를 숫자로 지정
import pandas as pd

# 데이터 프레임 생성
list2 = [
    ['홍길순', '여자', 22, 162],
    ['박철수', '남자', 30, 175],
    ['김영희', '여자', 28, 163],
    ['이철호', '남자', 35, 178],
    ['한지민', '여자', 24, 170],
    ['최영식', '남자', 29, 181],
    ['이수진', '여자', 23, 168],
    ['강호중', '남자', 32, 176],
    ['도봉순', '여자', 24, 172]
]

# 열 이름 리스트 만들기
col_name = ['이름','성별','나이','키']

# 데이터 프레임 생성
df = pd.DataFrame(list2, columns=col_name)

# 출력
df
```

```python
# 키가 180보다 큰 사람만 조회
df[df['키'] > 180]
```

## 조건식에 isin()함수,and기호(&),or(|),str.contains()함수를 사용하여 더 복잡한 조건으로 조회 할 수도 있음

```python
#나이가 21세 또는 23세인 사람을 조회
df[df['나이'].isin([21,23])]
```

```python
#성별이 여자이면서 키가 160보다 큰 사람 조회
# and = &
df[(df['성별'] == '여자' )& (df['키'] > 160)]
```

```python
#나이가 28세 이상이거나 성별이 남자인 사람을 조회
df[(df['나이'] >= 28) | (df['성별'] == '남자')]
```

```python
# 이름에 순이 들어가는 사람들 조회
df[df['이름'].str.contains('순')]
```

```python
# 숫자형 데이터만 통계를 구함
df.describe()
```

```python
# 인덱스 변경
df.set_index('이름',inplace=True) # inplace=True는 원본 데이터프레임에 적용

df
```

```python
df.head()
```

```python
# 열 추가 : 보너스 열 추가
# 나이에 1000을 곱한 값으로 지정

df['보너스'] = df['나이'] * 1000

df
```

```python
# 열 삭제
# drop , del
df.drop('보너스',axis=1,inplace=True)

df
```

```python
# del을 사용한 삭제

del df['보너스']

df
```

```python
# 기본 인덱스로 변경

df.reset_index(inplace = True)

df
```

```python
# 성별을 기준으로 조회하여 키와 평균과 표준편차 계산
df.describe()

# 평균 mean() 표준편차 std()
```

```python
# 성별의 평균 키
# 성별을 기준으로 그룹화 하여 그룹별 키의 평균을 계산
# as_index = False 데이터프레임으로 출력 됌
mean_by_gender = df.groupby(by=['성별'], as_index = False)['키'].mean()

# 열 이름 바꾸기
mean_by_gender.rename(columns={'키':'평균키'}, inplace=True)

mean_by_gender

```

```python
# 성별의 키의 표준편차
std_by_gender = df.groupby(by=['성별'], as_index = False).std()

# 열 이름 바꾸기
std_by_gender.rename(columns={'키':'키의 표준편차'}, inplace=True)

std_by_gender
```

```python
# 두 데이터프레임 합치기
new_df = pd.merge(mean_by_gender,std_by_gender)

new_df
```

```python
import pandas as pd

# 데이터 프레임
col_name = ['과목번호', '과목명', '강의실', '시간수']
list1 = [
    ['C1', '파이썬', '303호', 3],
    ['C2', '국어', '201호', 3],
    ['C3', '영어', '203호', 3],
    ['C4', '한국사', '201호', 2],
    ['C5', '과학', '401호', 3],
    ['C6', '사회', '201호', 2]
]
```

```python
# 시간표 데이터를 데이터프레임 객체 df로 변환하여 csv 파일로 저장
timeTable_df = pd.DataFrame(list1,columns=col_name)

# CSV 파일로 저장
timeTable_df.to_csv('./timetable.csv',header=True,index = False,encoding="utf-8")
```
